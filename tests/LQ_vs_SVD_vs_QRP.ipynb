```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
lq = {i:pd.read_csv(f'macaulayreduction/lq/dim{i}_power_randn.csv',index_col=0) for i in [2,3,4,5]}
svd = {i:pd.read_csv(f'macaulayreduction/svd/dim{i}_power_randn.csv',index_col=0) for i in [2,3,4,5]}
qrp = {i:pd.read_csv(f'macaulayreduction/qrp/dim{i}_power_randn.csv',index_col=0) for i in [2,3,4,5]}
```

```python
dims = np.arange(2, 6)
fig, ax = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(6, 4)


# Average Residuals
avg_res_lq = [lq[i]['avgres'][4] for i in dims]
avg_res_svd = [svd[i]['avgres'][4] for i in dims]
avg_res_qrp = [qrp[i]['avgres'][4] for i in dims]


ax[0].semilogy(dims, avg_res_lq, label='LQ')
ax[0].semilogy(dims, avg_res_svd, '--', label='SVD')
ax[0].semilogy(dims, avg_res_qrp, '-.', label='QRP')
ax[0].set_title("Average Residual")
ax[0].set_xlabel('Dimension')
ax[0].set_xticks([2, 3, 4, 5])
ax[0].legend()

# Max Eigenvalue Condition Number

mac_cond_lq = [lq[i]['avgcondeig'][4] for i in dims]
mac_cond_svd = [svd[i]['avgcondeig'][4] for i in dims]
mac_cond_qrp = [qrp[i]['avgcondeig'][4] for i in dims]

ax[1].semilogy(dims, mac_cond_lq, label='LQ')
ax[1].semilogy(dims, mac_cond_svd, '--', label='SVD')
ax[1].semilogy(dims, mac_cond_qrp, '-.', label='QRP')
ax[1].set_title("Average Condition Number")
ax[1].set_xlabel('Dimension')
ax[1].set_xticks([2, 3, 4, 5])

plt.suptitle("Degree 4 Polynomials")
plt.tight_layout()
plt.savefig("Degree_4_LQ_vs_QRP.pdf", dpi=300)
```

# Dimension 2

```python
dims = lq[2].index
fig = plt.figure(figsize=(12,8))

ax = fig.add_subplot(231)
ax.semilogy(dims,lq[2]['maxres'],label='LQ')
ax.semilogy(dims,svd[2]['maxres'],'--',label='SVD')
ax.semilogy(dims,qrp[2]['maxres'],'-.',label='QRP')
ax.set_title("Maximum Residual")
ax.set_xlabel('degree')
ax.legend()

ax = fig.add_subplot(232)
ax.semilogy(dims,lq[2]['avgres'],label='LQ')
ax.semilogy(dims,svd[2]['avgres'],'--',label='SVD')
ax.semilogy(dims,qrp[2]['avgres'],'-.',label='QRP')
ax.set_title("Average Residual")
ax.set_xlabel('degree')
ax.legend()

ax = fig.add_subplot(233)
ax.semilogy(dims,lq[2]['macaulay_cond_2'],label='LQ')
ax.semilogy(dims,svd[2]['macaulay_cond_2'],'--',label='SVD')
ax.semilogy(dims,qrp[2]['macaulay_cond_2'],'-.',label='QRP')
ax.set_title("Final Macaulay Condition Number")
ax.set_xlabel('degree')
ax.legend()

ax = fig.add_subplot(234)
ax.semilogy(dims,lq[2]['maxcondeig'],label='LQ')
ax.semilogy(dims,svd[2]['maxcondeig'],'--',label='SVD')
ax.semilogy(dims,qrp[2]['maxcondeig'],'-.',label='QRP')
ax.set_title("Maximum Eigenvalue Condition Number")
ax.set_xlabel('degree')
ax.legend()

ax = fig.add_subplot(235)
ax.semilogy(dims,lq[2]['avgcondeig'],label='LQ')
ax.semilogy(dims,svd[2]['avgcondeig'],'--',label='SVD')
ax.semilogy(dims,qrp[2]['avgcondeig'],'-.',label='QRP')
ax.set_title("Average Eigenvalue Condition Number")
ax.set_xlabel('degree')
ax.legend()

ax = fig.add_subplot(236)
ax.semilogy(dims,lq[2]['time'],label='LQ')
ax.semilogy(dims,svd[2]['time'],'--',label='SVD')
ax.semilogy(dims,qrp[2]['time'],'-.',label='QRP')
ax.set_title("Average Computation Time")
ax.set_xlabel('degree')
ax.set_ylabel('seconds')
ax.legend()

fig.suptitle("Dimension 2",fontsize='xx-large')
plt.tight_layout()
fig.subplots_adjust(top=0.9)
plt.savefig("2D_Solver_Comparison.pdf", dpi=300)
plt.show()
```
